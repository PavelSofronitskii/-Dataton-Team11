```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from time import sleep
```

```python
url = 'https://career.habr.com/vacancies?page=1&q=data&type=all'
```

```python
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.text, 'lxml')
```

```python
soup
```

```python
vacancy_name = soup.find('div', class_='vacancy-card__inner').find('a', class_='vacancy-card__title-link').text
vacancy_name
```

```python
company_name = soup.find('div', class_='vacancy-card__inner').find('a', class_='link-comp link-comp--appearance-dark').text
company_name
```

```python
salary = soup.find('div', class_='vacancy-card__inner').find('a', class_='basic-salary')
salary
```

```python
date = soup.find('div', class_='vacancy-card__inner').find('time', class_='basic-date').text
date
```

```python
job_type = []

job_types = soup.find('div', class_='vacancy-card__meta').find_all('span', class_='preserve-line')

for t in job_types:
    job_type.append(t.text)
    
job_type
```

```python
skill = []

skills = soup.find('div', class_='vacancy-card__inner').find('div', class_='vacancy-card__skills')

for s in skills:
    if s != ', ' and ' • ' not in s:
        skill.append(s.text)
    
skill
```

```python
data = []

vacancies = soup.findAll('div', class_='vacancy-card__inner')

for vacancy in vacancies:
    vacancy_name = vacancy.find('a', class_='vacancy-card__title-link').text
    company_name = vacancy.find('a', class_='link-comp link-comp--appearance-dark').text
    salary = vacancy.find('a', class_='basic-salary')
    date = soup.find('div', class_='vacancy-card__inner').find('time', class_='basic-date').text
        
    job_type = []
    job_types = soup.find('div', class_='vacancy-card__meta').find_all('span', class_='preserve-line')
    for t in job_types:
        job_type.append(t.text)
    
    skill = []
    skills = soup.find('div', class_='vacancy-card__inner').find('div', class_='vacancy-card__skills')
    for s in skills:
        if s != ', ' and ' • ' not in s:
            skill.append(s.text)
    
    
    data.append([vacancy_name, company_name,salary, date, job_type, skill])
```

```python
data
```

```python
df = []

for p in range(1,20):
    print(p)
    
    url = f'https://career.habr.com/vacancies?page={p}&q=data&type=all'
    r = requests.get(url)
    sleep(3)
    soup = BeautifulSoup(r.text, 'lxml')
    
    vacancies = soup.findAll('div', class_='vacancy-card__inner')
    
    for vacancy in vacancies:
        vacancy_name = vacancy.find('a', class_='vacancy-card__title-link').text
        company_name = vacancy.find('a', class_='link-comp link-comp--appearance-dark').text
        salary = vacancy.find('a', class_='basic-salary')
        date = soup.find('div', class_='vacancy-card__inner').find('time', class_='basic-date').text
        
        job_type = []
        job_types = soup.find('div', class_='vacancy-card__meta').find_all('span', class_='preserve-line')
        for t in job_types:
            job_type.append(t.text)

        skill = []
        skills = soup.find('div', class_='vacancy-card__inner').find('div', class_='vacancy-card__skills')
        for s in skills:
            if s != ', ' and ' • ' not in s:
                skill.append(s.text)


        df.append([vacancy_name, company_name,salary, date, job_type, skill])
```
